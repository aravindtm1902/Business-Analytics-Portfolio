```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.api as sn 
sns.set()
```

```python
data=pd.read_csv(r"C:\Users\tmara\Downloads\1.03.+Dummies.csv")
```

```python
data
```

```python
data1=data.copy()
```

```python
data1['Attendance']= data1['Attendance'].map({'Yes':1,'No':0})
```

```python
data1
```

```python
data1.describe()
```

```python
y=data1['GPA']
x1=data1[['SAT','Attendance']]
```

```python
x=sn.add_constant(x1)
results=sn.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data1['SAT'],y)
yhat_yes=0.8665+0.0014*data1['SAT']
yhat_no=0.639+0.0014*data1['SAT']
fig=plt.plot(data1['SAT'],yhat_no,lw=2,c='orange')
fig=plt.plot(data1['SAT'],yhat_yes,lw=2,c='green')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```
