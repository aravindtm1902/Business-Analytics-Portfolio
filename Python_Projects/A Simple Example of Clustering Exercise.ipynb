## Import the relevant libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
data=pd.read_csv(r"C:\Users\tmara\Downloads\Countries-exercise.csv")
data
```

## Plot the data

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

```python
x=data.iloc[:,1:3]
x
```

## Clustering

```python
kmeans= KMeans(7)
```

```python
kmeans.fit(x)
```

### Clustering Resutls

```python
identified_clusters=kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters=data.copy()
data_with_clusters['Clusters']=identified_clusters
data_with_clusters
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_with_clusters['Clusters'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
